```r
# Función para corregir una cadena de texto
corregir_texto <- function(texto) {
  # Cargar el diccionario de español
  diccionario <- hunspell_init("es_ES")

  # Divide el texto en palabras
  palabras <- unlist(strsplit(texto, " "))

  # Convierte las palabras a minúsculas
  palabras_minusculas <- tolower(palabras)

  # Corrige las palabras y reemplaza las incorrectas
  palabras_corregidas <- sapply(palabras_minusculas, function(palabra) {
    correccion <- hunspell_suggest(diccionario, palabra)
    if (length(correccion) > 0) {
      return(correccion[1])  # Utiliza la primera sugerencia si hay correcciones
    } else {
      return(palabra)  # Mantén la palabra original si no hay correcciones
    }
  })

  # Une las palabras corregidas en un texto nuevamente
  texto_corregido <- paste(palabras_corregidas, collapse = " ")

  # Cierra el diccionario
  hunspell_uninit(diccionario)

  return(texto_corregido)
}

# Aplica la corrección a la columna 'TITULO' y crea la columna 'descripcion'
data$descripcion <- sapply(data$TITULO, corregir_texto)

# Imprime las primeras 10 observaciones de las columnas 'TITULO' y 'descripcion'
head(data[c('TITULO', 'descripcion')], 10)
```

```r
# Instalar la biblioteca udpipe si aún no está instalada
# install.packages("udpipe")

# Cargar la biblioteca udpipe
library(udpipe)
```

```r
# Descargar el modelo pre-entrenado de español de UDPIPE
modelo_udpipe <- udpipe_download_model(language = "spanish")

# Cargar el modelo de español de UDPIPE
ud_model <- udpipe_model(modelo_udpipe$file_model)

# Tokeniza las palabras en cada descripción y obtén sus lemas y partes del discurso (POS)
data <- as.data.frame(data)  # Asegurarse de que 'data' sea un data frame
data_tokenizado <- udpipe_annotate(ud_model, x = data$descripcion)
data_lemmas_pos <- as.data.frame(data_tokenizado)

# Filtra las palabras para quedarte con verbos (VERB) y sustantivos (NOUN)
filter_verbs_nouns <- function(lemmas_pos_list) {
  filtered_words <- lemmas_pos_list$lemma[lemmas_pos_list$upos %in% c("VERB", "NOUN", "ADJ")]
  return(tolower(filtered_words))
}

data$descripcion_filtered <- sapply(data_lemmas_pos$lemma, filter_verbs_nouns)

# Cuenta la frecuencia de cada palabra filtrada
word_counts <- table(unlist(strsplit(data$descripcion_filtered, " ")))

# Ordena las palabras por frecuencia
most_common_words <- sort(word_counts, decreasing = TRUE)

# Imprime las palabras más comunes y sus frecuencias
head(most_common_words, 10)  # Puedes cambiar el número para obtener más o menos palabras
```

```r
# Instala y carga las bibliotecas tm y wordcloud si aún no están instaladas
# install.packages("tm")
# install.packages("wordcloud")
library(tm)
library(wordcloud)

# Convierte las frecuencias en un objeto de tipo 'tm' (DocumentTermMatrix)
corpus <- Corpus(VectorSource(names(word_counts)))
dtm <- DocumentTermMatrix(corpus, control = list(wordLengths = c(1, Inf)))

# Convierte el 'dtm' en una matriz de términos
matrix_terms <- as.matrix(dtm)

# Crea un objeto de frecuencia de palabras
word_freq <- sort(rowSums(matrix_terms), decreasing = TRUE)

# Crea la nube de palabras
wordcloud(words = names(word_freq), freq = word_freq, scale = c(3, 0.5), 
          min.freq = 1, colors = brewer.pal(8, "Dark2"))

# Guarda la imagen como un archivo PNG
png(filename = "wordcloud.png", width = 800, height = 400)
wordcloud(words = names(word_freq), freq = word_freq, scale = c(3, 0.5), 
          min.freq = 1, colors = brewer.pal(8, "Dark2"))
dev.off()
```

```r
# Define la función para calcular el monto total
calcular_monto_total <- function(row) {
  if (row$MONEDA == 'DOLARES') {
    monto_total <- (as.numeric(row$MONTO_RNR) + as.numeric(row$MONTO_FINANCIERO) + as.numeric(row$MONTO_NO_FINANCIERO)) * 3.6
  } else if (row$MONEDA == 'SOLES') {
    monto_total <- as.numeric(row$MONTO_RNR) + as.numeric(row$MONTO_FINANCIERO) + as.numeric(row$MONTO_NO_FINANCIERO)
  } else {
    monto_total <- 0.0
  }
  return(round(monto_total, 2))  # Redondear a dos decimales
}

# Aplica la función a cada fila del DataFrame y crea la nueva columna 'Monto total soles'
data$`Monto total soles` <- apply(data, 1, calcular_monto_total)

# Muestra el DataFrame con la nueva columna
print(data[, c('MONEDA', 'MONTO_RNR', 'MONTO_FINANCIERO', 'MONTO_NO_FINANCIERO', 'Monto total soles')])
```
